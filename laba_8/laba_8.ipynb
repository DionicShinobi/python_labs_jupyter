```python
print("Віріант №5")
print("Завдання(1) №2")
import pandas as pd

file_path = "NationalNames.csv"
data = pd.read_csv(file_path)

last_8_rows = data.tail(8)
print(last_8_rows)
```

```python
print("Завдання(2) №4")

numerical_summary = data.describe(include=[int, float])  
print( numerical_summary)
```

```python
print("Завдання(3) №6")

unique_female_names = data[data['Gender'] == 'F']['Name'].nunique()
unique_male_names = data[data['Gender'] == 'M']['Name'].nunique()

print(f"Кількість унікальних жіночих імен: {unique_female_names}")
print(f"Кількість унікальних чоловічих імен: {unique_male_names}")
```

```python
print("Завдання(4) №7")

male_names_2010 = data[(data['Year'] == 2010) & (data['Gender'] == 'M')]

top_5_male_names = male_names_2010.sort_values(by='Count', ascending=False).head(5)

print(top_5_male_names[['Name', 'Count']])
```

```python
print("Завдання(5) №9")

min_count = data['Count'].min()

min_count_records = data[data['Count'] == min_count]
count_of_min_count_records = min_count_records.shape[0]

print(f"Мінімальне значення Count: {min_count}")
print(f"Кількість записів із мінімальним Count: {count_of_min_count_records}")
```

```python
print("Завдання(6) №10")

unique_names_per_year = data.groupby('Year')['Name'].nunique()

print("Year  Name")
for year, count in unique_names_per_year.items():
    print(f"{year}: {count}")
```

```python
print("Завдання(7) №15")

total_births_per_year = data.groupby('Year')['Count'].sum()

print("Year Total Births")
for year, total in total_births_per_year.items():
    print(f"{year} {total}")
```

```python
print("Завдання(8) №16")

max_birth_year = total_births_per_year.idxmax()

print(f"Рік з найбільшим числом народжень: {max_birth_year}")
```

```python
print("Завдання(9) №17")

births_per_year_gender = data.groupby(['Year', 'Gender'])['Count'].sum().unstack(fill_value=0)

print("Year   F     M")
for year, row in births_per_year_gender.iterrows():
    print(f"{year} {row['F']} {row['M']}")
```

```python
print("Завдання(10) №18")

births_per_year_gender = data.groupby(['Year', 'Gender'])['Count'].sum().unstack(fill_value=0)

count_years_more_girls = (births_per_year_gender['F'] > births_per_year_gender['M']).sum()

print(f"Кількість років, коли дівчаток народжувалось більше, ніж хлопчиків: {count_years_more_girls}")
```

```python
print("Завдання(11) №20")

gender_name_counts = data.groupby(['Name', 'Gender']).size().unstack(fill_value=0)

neutral_names = gender_name_counts[(gender_name_counts['F'] > 0) & (gender_name_counts['M'] > 0)]

count_neutral_names = len(neutral_names)

print(f"Кількість гендерно-нейтральних імен (які є для дівчат та хлопців): {count_neutral_names}")
```

```python
print("Завдання(12) №21")
barbara_boys = data[(data['Name'] == 'Barbara') & (data['Gender'] == 'M')]

count_barbara_boys = barbara_boys['Count'].sum()

print(f"Кількість разів, коли хлопчиків називали Barbara: {count_barbara_boys}")
```

```python
print("Завдання(13) №22")

num_years = data['Year'].nunique()

print(f"Спостереження проводилося {num_years} разів")
```

```python
print("Завдання(14) №23")

male_names = data[data['Gender'] == 'M']['Name'].unique()
female_names = data[data['Gender'] == 'F']['Name'].unique()

neutral_names = set(male_names).intersection(set(female_names))

neutral_names_count = data[data['Name'].isin(neutral_names)]

neutral_names_count = neutral_names_count.groupby('Name', as_index=False)['Count'].sum()

neutral_names_count = neutral_names_count.sort_values(by='Count', ascending=False)

result = neutral_names_count[['Name']].head(16) 
print(result.to_string(index=False)) 
```

```python
print("Завдання(15) №24")

print("Завдання: Найпопулярніше серед непопулярних імен")
```

```python
print("Завдання(16) №25")

import matplotlib.pyplot as plt

john_data = data[data['Name'] == 'John']
mary_data = data[data['Name'] == 'Mary']

john_yearly = john_data.groupby('Year')['Count'].sum()
mary_yearly = mary_data.groupby('Year')['Count'].sum()

plt.figure(figsize=(10, 6))

plt.plot(john_yearly.index, john_yearly.values, color='blue', label='John', linestyle='-', linewidth=2)

plt.plot(mary_yearly.index, mary_yearly.values, color='orange', label='Mary', linestyle='-', linewidth=2)

plt.title("Розподіл кількості імен 'John' та 'Mary' по роках")
plt.xlabel("Рік")
plt.ylabel("Кількість народжень")
plt.legend()
plt.grid(True)

# Показуємо графік
plt.tight_layout()
plt.show()
```

```python
print("Завдання(17) №26")
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("NationalNames.csv")

john_female_data = data[(data['Name'] == 'John') & (data['Gender'] == 'F')]
mary_male_data = data[(data['Name'] == 'Mary') & (data['Gender'] == 'M')]

john_female_yearly = john_female_data.groupby('Year')['Count'].sum()
mary_male_yearly = mary_male_data.groupby('Year')['Count'].sum()

plt.figure(figsize=(10, 6))

plt.plot(john_female_yearly.index, john_female_yearly.values, color='purple', label='John (Female)', linestyle='-', linewidth=2)

plt.plot(mary_male_yearly.index, mary_male_yearly.values, color='green', label='Mary (Male)', linestyle='-', linewidth=2)

plt.title("Розподіл кількості жіночих імен 'John' та чоловічих імен 'Mary' по роках")
plt.xlabel("Рік")
plt.ylabel("Кількість народжень")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
print("Завдання(18) №27")

grouped_data = data.groupby(['Year', 'Name'])['Count'].sum().reset_index()

most_popular_names = grouped_data.loc[grouped_data.groupby('Year')['Count'].idxmax()]

print("Year   Most P.N.   Count")
for _, row in most_popular_names.iterrows():
    print(f"{row['Year']}   {row['Name']}        {row['Count']}")
```
